```python
import pandas as pd
import numpy as np
import scipy as cp
import matplotlib.pyplot as plt
import math
print("Set up")
```

```python
r = 130
h = 1378
l = 1757
d = 0.73
s = math.pi * d * 2 / 4
data = pd.read_csv("../data/ler.csv")
data
```

```python
def dlen(n):
    return n*r/2/h

for i in range(1, 4):
    data["l{}".format(i)] = dlen(data["n{}".format(i)])
data["l1"] -= data["l1"][0]
data["l2"] -= data["l2"][1]
data["l3"] -= data["l3"][0]

for i in range(1, 4):
    for j in range(1, 10):
        data["m{}".format(i)][j] += data["m{}".format(i)][j-1]
data["m2"] -= data["m2"][1]
        
data
```

```python
del data["n2"][0]
del data["m2"][0]
del data["l2"][0]
del data["n3"][9]
del data["m3"][9]
del data["l3"][9]
```

```python
plt.title("Зависимость удлинения от нагрузки")
plt.xlabel("$\Delta l$, мм")
plt.ylabel("$F$, Н")
plt.grid(True)
for i in range(1, 4):
    plt.plot(data["l{}".format(i)], data["m{}".format(i)]*10/1000, "{}.".format(['r', 'g', 'b'][i-1]))
plt.savefig("../img/dirty.png", dpi=400)
```

```python
del data["m3"][7]
del data["l3"][7]
del data["n3"][7]
#del data["m2"][0]
#del data["l3"][0]
```

```python
plt.title("Зависимость удлинения от нагрузки")
plt.xlabel("$\Delta l$, мм")
plt.ylabel("$F$, Н")
plt.grid(True)
for i in range(1, 4):
    plt.plot(data["l{}".format(i)], data["m{}".format(i)]*10/1000, "{}.".format(['r', 'g', 'b'][i-1]))
```

```python
m = np.concatenate(tuple(data["m{}".format(i)].to_numpy() for i in range(1, 4)), axis=0) / 1000 * 10
ll = np.concatenate(tuple(data["l{}".format(i)].to_numpy() for i in range(1, 4)), axis=0)
```

```python
k = np.mean(m * ll) / np.mean(ll**2)
```

```python
n = len(m-3)
err = math.sqrt(1/n/(n-1)*sum([(x - k)**2 for x in m[np.where(m != 0)]/ll[np.where(ll != 0)]]))
```

```python
print("{} +/- {}".format(k, err))
```

```python
plt.title("Зависимость удлинения от нагрузки")
plt.xlabel("$\Delta l$, мм")
plt.ylabel("$F$, Н")
plt.grid(True)
for i in range(1, 4):
    plt.plot(data["l{}".format(i)], data["m{}".format(i)]*10/1000, "{}.".format(['r', 'g', 'b'][i-1]))
plt.axline((0, 0), (4, 4*k))
plt.savefig("../img/clean.png", dpi=400)
```

```python
print("E = {:e} +/- {:e}".format(k * 1e3 * l * 1e-2 / (s * 1e-6), err  *1e3 * l * 1e-2 / (s * 1e-6)))
```
