```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
from datetime import datetime
print("Done")
```

```python
def get_data(filename):
    data = pd.read_csv(filename)
    data = data.drop(columns=["Unnamed: 2"])
    data["Time"] = data["Time"].apply(lambda x: datetime.strptime(x, '%H : %M : %S'))
    data["Time"] = data["Time"] - data["Time"][0]
    data["Time"] = data["Time"].apply(lambda dt: dt.seconds)
    return data
```

```python
data1 = get_data("../data/data1.csv")
data1 = data1.rename(columns={"Value": "R"})

data2 = get_data("../data/data2.csv")
data2 = data2.rename(columns={"Value": "Tk"})

data = pd.concat([data1, data2.drop(columns=["Time"])], axis=1)
data
```

```python
data["T"] = 14.583955001619313455 * data["R"] + 39.35514018691588785 - 273
data
```

```python
cool_empty_start = 482
cool_empty_end = 680
heat_empty_start = 1050
heat_empty_end = 2650
cool_fe_start = 2950
cool_fe_end = 3100
heat_fe_start = 3350
heat_fe_end = 4800
cool_al_start = 4800
cool_al_end = 5000
heat_al_start = 5200
heat_al_end = 6300
```

```python
def line(x):
    plt.axvline(x=x, linewidth=0.5, color='green')
    
plt.xlabel("$t$, с")
plt.ylabel("$T$, $^\circ$C")
plt.plot(data["Time"], data["Tk"])
plt.plot(data["Time"], data["T"], '--')
line(cool_empty_start)
line(cool_empty_end)
line(heat_empty_start)
line(heat_empty_end)
line(cool_fe_start)
line(cool_fe_end)
line(heat_fe_start)
line(heat_fe_end)
line(cool_al_start)
line(cool_al_end)
line(heat_al_start)
line(heat_al_end)
plt.savefig("fig.png", dpi=400)
```

```python
d = data.iloc[cool_empty_start:cool_empty_end]
plt.plot(d["Time"], d["T"], '--')
```
