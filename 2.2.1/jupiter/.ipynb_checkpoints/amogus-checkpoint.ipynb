```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
from datetime import datetime
print("Done")

plt.style.use('/home/leonid/github.com/LeonidPilyugin/mpl-style/simple.mplstyle')
```

```python
def get_tau(path):
    plt.cla()
    data = pd.read_csv(path)
    y = np.log(data["V"])
    x = data["t"]
    i = int(path[-5])
    if i == 1:
        x = x[200:]
        y = y[200:]
    elif i == 2:
        x = x[300:]
        y = y[300:]
    elif i == 3:
        x = x[400:]
        y = y[400:]
    elif i == 4:
        x = x[500:]
        y = y[500:]
    elif i == 5:
        x = x[600:]
        y = y[600:]
    elif i == 6:
        x = x[600:]
        y = y[600:]
    
    pp, V = np.polyfit(x, y, 1, cov=True)
    k, b = pp
    k_err, b_err = np.sqrt(V[0][0]), np.sqrt(V[1][1])
    
    plt.plot(data["t"], np.log(data["V"]), '.')
    plt.axline([min(x), min(x) * k + b], [max(x), max(x) * k + b], color='r')
    plt.grid()
    plt.xlabel("$t$, с")
    plt.ylabel("ln $V$")
    plt.title(f"$p = {([51.5,102.9,147,205.9,257.3,297.8])[i-1]}$ торр")
    plt.savefig(path.split("/")[-1].replace(".csv", ".png"), dpi=400)
    return 1/k, 1/k*(k_err/k)
```

```python
tp = [(*get_tau(f"../data/data{i}.csv"), ([51.5,102.9,147,205.9,257.3,297.8])[i-1]) for i in range(1,7)]
tp
```

```python
v = 800e-6
v_err = 5e-6
ls = 1500
ls_err = 100
```

```python
dp = [(-v*ls/2/t, -v*ls/2/t * (t_err / t + v_err / v + ls_err / ls), p * 133.3) for t, t_err, p in tp]
dp
```

```python
x = [1/p for d, d_err, p in dp]
y = [d for d, d_err, p in dp]
pp, V = np.polyfit(x + [0], y + [0], 1, cov=True, w=[0, 1, 1, 1, 1, 1, 1e9])
k, b = pp
k_err, b_err = np.sqrt(V[0][0]), np.sqrt(V[1][1])
plt.plot(x, y, '.')
plt.xlabel("$1/p$, Па$^{-1}$")
plt.ylabel("D, м$^2$/с")
plt.grid()
plt.axline([min(x), min(x) * k + b], [max(x), max(x) * k + b], color='r')
plt.plot(0, b, 'g.')
plt.savefig("total.png", dpi=400)
```
