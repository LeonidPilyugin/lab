```python
R =  20000 * u("ohm")
C = 20e-6 * u("F")
R0 = 0.22 * u("ohm")
names = ["феррит", "Fe-Si", "Fe-Ni"]
N0 = [42, 25, 20]
NU = [400, 250, 300]
S = [3 * u("cm**2"), 2 * u("cm**2"), 0.76 * u("cm**2")]
L = [25 * u("cm"), 11 * u("cm"), 13.3 * u("cm")]

kH = [N0[i]/R0/L[i] for i in range(3)]
kB = [R*C/S[i]/NU[i] for i in range(3)]

for i in range(3):
    kB[i].ito(u("tesla/volt"))
    kH[i].ito(u("ampere/meter/volt"))

_2xs = [317.5 * u("mV"), 162 * u("mV"), 68 * u("mV")]
_2ys = [124 * u("mV"), 80 * u("mV"), 162 * u("mV")]
_2xc = [50 * u("mV"), 66 * u("mV"), 64 * u("mV")]
_2yc = [46 * u("mV"), 44 * u("mV"), 162 * u("mV")]

Bs = [kB[i] * _2ys[i] / 2 for i in range(3)]
Bc = [kB[i] * _2yc[i] / 2 for i in range(3)]
Hs = [kH[i] * _2xs[i] / 2 for i in range(3)]
Hc = [kH[i] * _2xc[i] / 2 for i in range(3)]
```
