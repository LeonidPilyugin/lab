```python
def get_tau(path):
    plt.cla()
    data = pd.read_csv(path)
    y = -np.log(data["V"])
    x = data["t"]
    i = int(path[-5])
    if i == 1:
        print("lol")
        x = x[:200]
        y = y[:200]
    elif i == 2:
        x = x[:300]
        y = y[:300]
    elif i == 3:
        x = x[:400]
        y = y[:400]
    elif i == 4:
        x = x[:500]
        y = y[:500]
    elif i == 5:
        x = x[:600]
        y = y[:600]
    elif i == 6:
        x = x[:600]
        y = y[:600]
    
    pp, V = np.polyfit(x, y, 1, cov=True)
    k, b = pp
    k_err, b_err = np.sqrt(V[0][0]), np.sqrt(V[1][1])
    
    plt.plot(data["t"], -np.log(data["V"]), '.')
    plt.axline([min(x), min(x) * k + b], [max(x), max(x) * k + b], color='r')
    plt.grid()
    plt.xlabel("$t$, —Å")
    plt.ylabel("ln $V$")
    plt.savefig(path.split("/")[-1].replace(".csv", ".png"), dpi=400)
    return k, k_err
```

```python
def eta(path, l, R, R_err):
    plt.cla()
    data = pd.read_csv(path)
    data["Q"] /= 1e3
    data["p"] = data["h"] * data["a"] * 9.80665
    
    x = data["p"]
    y = data["Q"]
    pp, V = np.polyfit(x, y, 1, cov=True)
    k, b = pp
    k_err, b_err = np.sqrt(V[0][0]), np.sqrt(V[1][1])
    plt.axline([min(x), min(x) * k  *1000 + b * 1000], [max(x), max(x) * k * 1000 + b * 1000], color='r')
    plt.plot(x, y * 1000, '.')
    plt.xlabel("$\Delta P$, –ü–∞")
    plt.ylabel("$Q$, –ª$/$—Å")
    plt.grid()
    plt.title(f"Q(p) –¥–ª—è —Ç—Ä—É–±–∫–∏ {path[-5]}")
    plt.savefig(path.split("/")[-1].replace(".csv", ".png"), dpi=400)
    
    eta = np.pi * R ** 4 / 8 / k / l
    eta_err = eta * (4 * R_err / R + k_err / k)
    print(f"eta = {eta} +/- {eta_err}")
```

```python
eta("../data/data01.csv", 0.4, 5.1e-3 / 2, 0.05e-3 / 2)
eta("../data/data02.csv", 0.3, 3.0e-3 / 2, 0.1e-3 / 2)
eta("../data/data03.csv", 0.5, 3.95e-3 / 2, 0.05e-3 / 2)
```

```python
def distribution(path):
    plt.cla()
    data = pd.read_csv(path)
    data["p"] = data["h"] * data["a"] * 9.80665
    data["pp"] = data["p"].cumsum()
    data["x"] = data["l"][::-1].cumsum()
    
    x = data["x"]
    y = data["pp"]
    k, b = np.polyfit(x[:-2] if len(x) > 3 else x, y[:-2] if len(y) > 3 else y, 1)
    plt.axline([min(x), min(x) * k + b], [max(x), max(x) * k + b], color='r')
    
    plt.plot(data["x"], data["pp"], '.')
    plt.xlabel("$x$, —Å–º")
    plt.ylabel("$p-p_c$, –ü–∞")
    plt.grid()
    plt.title(f"P(x) –¥–ª—è —Ç—Ä—É–±–∫–∏ {path[-5]}")
    plt.savefig(path.split("/")[-1].replace(".csv", ".png"), dpi=400)
```

```python
distribution("../data/data11.csv")
distribution("../data/data12.csv")
distribution("../data/data13.csv")
```
