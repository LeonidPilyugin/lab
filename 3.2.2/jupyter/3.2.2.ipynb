# Texify primary data

```python
for df, filename in zip((data, achh1, achh2, fchh1, fchh2), ("data", "achh1", "achh2", "fchh1", "fchh2")):
    utils.totex(df, file=f"../tex/{filename}_primary_table.tex", columns={
        "C": "C",
        "x1": "x_1",
        "x2": "x_2",
        "phi": "\\Delta\\phi",
        "C_n": "C_n",
        "f": "\\nu",
        "U_C": "U_C",
        "E": "\\mathcal{E}",
        "Q": "Q",
        "R_Sigma": "R_{\Sigma}",
        "I": "I",
        "L": "L",
        "rho": "\\rho",
        "R_Smax": "R_{S_{max}}",
        "R_L": "R_L",
        "rel": "R_{S_{max}} / R_{\Sigma}"
    })
```

# Compute

```python
data.head()
```

```python
data["Q"] = data["U_C"] / data["E"]
data["R_Sigma"] =  ((data["f"] * 2 * pi * data["C_n"] * data["Q"]) ** -1).ito(u.Ω)
data["I"] = (data["E"] / data["R_Sigma"]).ito(u.A)
data["L"] = (data["R_Sigma"] * data["Q"] / data["f"] / 2 / pi).ito(u.μH)
data["rho"] = data["Q"] * data["R_Sigma"]
data["R_Smax"] = 1e-3 * data["rho"]
data["R_L"] = data["R_Sigma"] - data["R_Smax"]
data.head(6)
```
